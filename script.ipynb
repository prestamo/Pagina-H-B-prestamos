# [dbo].[ventasview]

```sql
/****** Object:  View [dbo].[ventasview]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-------- vistas de ventas
create view [dbo].[ventasview]
as
select v.num_documento as 'NumeroDocumento',v.serie as 'Numero de Serie',v.fecha,c.nombres +' ' + c.apellidos as 'Cliente',
v.total as 'Ganancia',v.estado from cliente c inner join venta v on c.cod_cliente=v.cod_cliente
GO
```

# [dbo].[abrir_caja]

```sql
/****** Object:  StoredProcedure [dbo].[abrir_caja]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[abrir_caja]
    @id_caja INT,
    @fecha DATE,
    @id_usuario INT,
    @fondo DECIMAL(18, 2) -- Fondo inicial que se va a registrar en el historial
AS
BEGIN
    -- Eliminar cualquier entrada existente que podría causar duplicación
    DELETE FROM EstadoCaja WHERE id_caja = @id_caja AND estado = 'Abierta';

    -- Insertar el registro de apertura de caja
    INSERT INTO EstadoCaja (id_caja, estado, fecha_apertura)
    VALUES (@id_caja, 'Abierta', @fecha);

    -- No se actualiza el fondo de la caja (se deja en 0), solo la apertura
    UPDATE Cajas
    SET estado = 'Abierta', 
        fecha_apertura = @fecha, 
        id_usuario = @id_usuario,
        fondo = 0 -- El fondo se mantiene en 0
    WHERE id_caja = @id_caja;

    -- Registrar en el Historial de Caja el fondo inicial ingresado
    INSERT INTO HistorialCaja (id_caja, fecha_apertura, fondo_inicial, id_usuario)
    VALUES (@id_caja, @fecha, @fondo, @id_usuario);
END
GO
```

# [dbo].[ActualizarEstadoCredito]

```sql
/****** Object:  StoredProcedure [dbo].[ActualizarEstadoCredito]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[ActualizarEstadoCredito]
    @id_credito INT
AS
BEGIN
    -- Verificar si el crédito está completamente pagado
    IF EXISTS (
        SELECT 1 
        FROM ClienteCredito 
        WHERE id_credito = @id_credito AND monto_adeudado = 0
    )
    BEGIN
        -- Actualizar el estado a 'Pagado'
        UPDATE ClienteCredito
        SET estado = 'Pagado'
        WHERE id_credito = @id_credito;
    END
END;
GO
```

# [dbo].[bajar_stock]

```sql
/****** Object:  StoredProcedure [dbo].[bajar_stock]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--- bajar el stop del producto cuando se reaiza una venta 

CREATE PROCEDURE [dbo].[bajar_stock]
    @cod_producto CHAR(10),
    @stock INT
AS
BEGIN
    -- Verificar si el producto existe
    IF EXISTS (SELECT 1 FROM producto WHERE cod_producto = @cod_producto)
    BEGIN
        -- Reducir el stock del producto
        UPDATE producto
        SET stock = stock - @stock
        WHERE cod_producto = @cod_producto;
    END
    ELSE
    BEGIN
        -- Manejar el caso donde el producto no existe
        RAISERROR ('El producto no existe.', 16, 1);
    END
END;
GO
```

# [dbo].[BuscarFacturasPorFecha]

```sql
/****** Object:  StoredProcedure [dbo].[BuscarFacturasPorFecha]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[BuscarFacturasPorFecha]
    @fecha DATE
AS
BEGIN
    -- Seleccionar las facturas en la fecha especificada
    SELECT *
    FROM Facturas
    WHERE CAST(fecha AS DATE) = @fecha;

    -- Seleccionar los detalles de todas las facturas de esa fecha
    SELECT DF.*
    FROM DetallesFactura DF
    INNER JOIN Facturas F ON DF.numero_factura = F.numero_factura
    WHERE CAST(F.fecha AS DATE) = @fecha;
END;
GO
```

# [dbo].[BuscarUltimaFactura]

```sql
/****** Object:  StoredProcedure [dbo].[BuscarUltimaFactura]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[BuscarUltimaFactura]
AS
BEGIN
    -- Obtener la última factura según la fecha más reciente
    SELECT TOP 1 *
    FROM Facturas
    ORDER BY fecha DESC;

    -- Obtener los detalles de la última factura
    SELECT DF.*
    FROM DetallesFactura DF
    INNER JOIN (
        SELECT TOP 1 numero_factura
        FROM Facturas
        ORDER BY fecha DESC
    ) UltimaFactura ON DF.numero_factura = UltimaFactura.numero_factura;
END;
GO
```

# [dbo].[eliminar_monto_caja]

```sql
/****** Object:  StoredProcedure [dbo].[eliminar_monto_caja]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[eliminar_monto_caja]
    @num_documento VARCHAR(50)
AS
BEGIN
    DECLARE @id_caja INT;
    DECLARE @monto_efectivo DECIMAL(18, 2);

    -- Obtener el monto total de efectivo de la venta
    SELECT @monto_efectivo = ISNULL(SUM(monto), 0)
    FROM Transacciones 
    WHERE num_documento = @num_documento AND tipo = 'Efectivo';

    -- Obtener el id de la caja desde la tabla Transacciones
    SELECT TOP 1 @id_caja = id_caja 
    FROM Transacciones 
    WHERE num_documento = @num_documento;

    -- Actualizar el fondo de la caja, restando el monto de efectivo
    IF @monto_efectivo > 0 AND @id_caja IS NOT NULL
    BEGIN
        UPDATE Cajas
        SET fondo = fondo - @monto_efectivo
        WHERE id_caja = @id_caja;
    END

    -- Eliminar las transacciones relacionadas con el num_documento especificado
    DELETE FROM Transacciones
    WHERE num_documento = @num_documento;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[eliminar_venta]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[eliminar_venta](
    @num_documento VARCHAR(50)
)
AS
BEGIN
    -- Declarar variables para los detalles
    DECLARE @cod_producto VARCHAR(10), @cantidad INT, @serie CHAR(4), @fecha DATE, @cod_cliente CHAR(10), @dni VARCHAR(10), @total DECIMAL(18,2), @estado CHAR(1);
    
    -- Recorremos los detalles de la venta y almacenamos la información en la tabla ventas_eliminadas
    DECLARE cur CURSOR FOR
    SELECT cod_producto, cantidad FROM detalleventa WHERE num_documento = @num_documento;
    
    OPEN cur;
    FETCH NEXT FROM cur INTO @cod_producto, @cantidad;
    
    WHILE @@FETCH_STATUS = 0
    BEGIN
        -- Insertar la información en la tabla ventas_eliminadas con detalles de producto
        INSERT INTO ventas_eliminadas (num_documento, serie, fecha, cod_cliente, dni, total, estado, cod_producto, nombre_producto, descripcion_producto, precio_compra, precio_venta, cantidad)
        SELECT 
            v.num_documento, 
            v.serie, 
            v.fecha, 
            v.cod_cliente, 
            v.dni, 
            v.total, 
            v.estado,
            d.cod_producto,
            p.nombre,
            p.descripcion_producto,
            p.precio_compra,
            p.precio_venta,
            d.cantidad
        FROM venta v
        INNER JOIN detalleventa d
            ON v.num_documento = d.num_documento
            AND v.serie = d.serie
        INNER JOIN producto p
            ON d.cod_producto = p.cod_producto
        WHERE v.num_documento = @num_documento;

        -- Devolver el stock al producto
        UPDATE producto
        SET stock = stock + @cantidad
        WHERE cod_producto = @cod_producto;

        FETCH NEXT FROM cur INTO @cod_producto, @cantidad;
    END;

    CLOSE cur;
    DEALLOCATE cur;
    
    -- Eliminar los detalles de la venta
    DELETE FROM detalleventa WHERE num_documento = @num_documento;
    DELETE FROM venta WHERE num_documento = @num_documento;
END;
GO
```

# [dbo].[generar_num_factura]

```sql
/****** Object:  StoredProcedure [dbo].[generar_num_factura]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[generar_num_factura]
    @num_documento VARCHAR(20) OUTPUT
AS
BEGIN
    DECLARE @nuevo_num_documento INT;
    -- Obtener el último número de documento
    SELECT @nuevo_num_documento = ISNULL(MAX(ultimo_num_documento), 0) FROM control_factura;
    -- Asignar el nuevo número de documento con el prefijo "FACTURA-"
    SET @num_documento = 'FACTURA-' + RIGHT('0000000000' + CAST(@nuevo_num_documento + 1 AS VARCHAR), 10);
    -- Actualizar el número en la tabla de control
    IF EXISTS (SELECT 1 FROM control_factura)
    BEGIN
        UPDATE control_factura SET ultimo_num_documento = @nuevo_num_documento + 1;
    END
    ELSE
    BEGIN
        INSERT INTO control_factura (ultimo_num_documento) VALUES (@nuevo_num_documento + 1);
    END
END;
GO
```

# [dbo].[insertar_usuario]

```sql
/****** Object:  StoredProcedure [dbo].[insertar_usuario]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- registrar usuario
CREATE PROCEDURE [dbo].[insertar_usuario]
(
    @cod_usuario CHAR(10) OUTPUT,  -- Parámetro de salida
    @usuario VARCHAR(40),
    @contraseña VARCHAR(40),
    @tipo_usuario VARCHAR(20)
)
AS
BEGIN
    -- Declaración de variables
    DECLARE @num INT = 0; -- Variable para almacenar la parte numérica del código de usuario
    DECLARE @nuevo_cod_usuario CHAR(10); -- Variable para generar el nuevo código de usuario

    -- Verifica si ya existen registros en la tabla usuario
    IF EXISTS (SELECT cod_usuario FROM usuario)
    BEGIN
        -- Obtén el número más alto actual, extrayendo solo la parte numérica después del prefijo 'USR'
        SELECT @num = MAX(CONVERT(INT, SUBSTRING(cod_usuario, 4, LEN(cod_usuario) - 3)))
        FROM usuario
        WHERE ISNUMERIC(SUBSTRING(cod_usuario, 4, LEN(cod_usuario) - 3)) = 1;

        -- Si @num es NULL (si no hay registros numéricos válidos), inicialízalo en 0
        IF @num IS NULL
        BEGIN
            SET @num = 0;
        END

        -- Genera el nuevo código de usuario con el formato 'USR' seguido de ceros
        SET @nuevo_cod_usuario = 'USER' + RIGHT('000000' + CONVERT(VARCHAR, @num + 1), 7);
    END
    ELSE
    BEGIN
        -- Si no hay registros, comienza con USR0000001
        SET @nuevo_cod_usuario = 'USER0000001';
    END;

    -- Asignar el nuevo código de usuario a la variable de salida
    SET @cod_usuario = @nuevo_cod_usuario;

    -- Inserta el nuevo usuario en la tabla
    INSERT INTO usuario (cod_usuario, usuario, contraseña, tipo_usuario, estado)
    VALUES (@cod_usuario, @usuario, @contraseña, @tipo_usuario, '1');
END;
GO
```

# [dbo].[mostrar_stock_productos]

```sql
/****** Object:  StoredProcedure [dbo].[mostrar_stock_productos]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- procedimiento para mostrar stop bajos y medios para llevar un control de la compra de mercansias
CREATE PROCEDURE [dbo].[mostrar_stock_productos]
AS
BEGIN
    SELECT 
        p.cod_producto, 
        p.nombre, 
        p.Descripcion_producto, 
        p.stock, 
        p.cod_categoria, 
        p.estado, 
        p.precio_compra, 
        p.precio_venta,
        pr.nombre AS nombre_proveedor,     -- Nombre del proveedor
        p.fecha_registro,                  -- Fecha de registro del producto
        CASE 
            WHEN p.stock <= 5 THEN 'BAJO' 
            WHEN p.stock BETWEEN 6 AND 9 THEN 'MEDIO' 
            ELSE 'ALTO' 
        END AS nivel_stock                 -- Nivel de stock calculado
    FROM 
        producto p
    JOIN 
        Proveedores pr ON p.id_proveedor = pr.id_proveedor
    WHERE 
        p.estado = '1';
END;
GO
```

# [dbo].[ObtenerResumenCajaPorFecha]

```sql
/****** Object:  StoredProcedure [dbo].[ObtenerResumenCajaPorFecha]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

---- nueva consulta obtimisada para el historial de caja
CREATE PROCEDURE [dbo].[ObtenerResumenCajaPorFecha]
    @FechaInicio DATE,
    @FechaFin DATE
AS
BEGIN
    WITH TransaccionesResumen AS (
        SELECT 
            u.nombre_usuario AS Usuario,
            c.nombre_caja AS NombreCaja,
            h.fondo_inicial AS FondoInicial,
            h.fecha_apertura AS FechaInicial,
            h.fecha_cierre AS FechaFinal,
            CASE WHEN t.tipo = 'Efectivo' THEN t.monto ELSE 0 END AS MontoEfectivo,
            CASE WHEN t.tipo = 'Deposito' THEN t.monto ELSE 0 END AS MontoDeposito,
            CASE WHEN t.tipo = 'Transferencia' THEN t.monto ELSE 0 END AS MontoTransferencia,
            -- Calcular el fondo acumulado sin incluir el fondo inicial
            SUM(CASE 
                    WHEN t.tipo = 'Efectivo' THEN t.monto 
                    WHEN t.tipo = 'Deposito' THEN t.monto 
                    WHEN t.tipo = 'Transferencia' THEN t.monto 
                    ELSE 0 
                END) OVER (PARTITION BY h.id_caja ORDER BY t.fecha ROWS UNBOUNDED PRECEDING) AS FondoTotal
        FROM 
            HistorialCaja h
        INNER JOIN 
            Cajas c ON h.id_caja = c.id_caja
        INNER JOIN 
            UsuariosLogueados u ON h.id_usuario = u.id
        INNER JOIN 
            Transacciones t ON h.id_caja = t.id_caja
        WHERE 
            h.fecha_apertura BETWEEN @FechaInicio AND @FechaFin
    )
    SELECT 
        Usuario,
        NombreCaja,
        FondoInicial,
        MontoEfectivo,
        MontoDeposito,
        MontoTransferencia,
        FondoTotal AS TotalEnCaja,
        FechaInicial,
        FechaFinal
    FROM 
        TransaccionesResumen;
END
GO
```

# [dbo].[RegistrarPago]

```sql
/****** Object:  StoredProcedure [dbo].[RegistrarPago]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[RegistrarPago]
    @id_credito INT,
    @monto_pago DECIMAL(18, 2)
AS
BEGIN
    BEGIN TRANSACTION;

    -- Insertar el pago
    INSERT INTO PagosCredito (id_credito, fecha_pago, monto_pago)
    VALUES (@id_credito, GETDATE(), @monto_pago);

    -- Actualizar el monto adeudado
    UPDATE ClienteCredito
    SET monto_adeudado = monto_adeudado - @monto_pago
    WHERE id_credito = @id_credito;

    -- Verificar si el crédito está pagado y actualizar el estado
    EXEC ActualizarEstadoCredito @id_credito;

    COMMIT TRANSACTION;
END;
GO
```

# [dbo].[sp_buscar_venta_por_numero_documento]

```sql
/****** Object:  StoredProcedure [dbo].[sp_buscar_venta_por_numero_documento]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
---------- procedimiento para buscar el factura con todos los productos de esa factura
CREATE PROCEDURE [dbo].[sp_buscar_venta_por_numero_documento]
    @NumeroDocumento NVARCHAR(50)
AS
BEGIN
    -- Declaramos la variable para contar los productos
    DECLARE @CantidadProductos INT;
    
    -- Buscamos los detalles de la venta por número de documento
    SELECT 
        v.num_documento AS [Número de Factura],
        v.serie AS [Serie],
        v.cod_cliente AS [Código Cliente],
        v.dni AS [DNI],
        d.cod_producto AS [Código Producto],
        p.nombre AS [Nombre Producto],
        p.descripcion_producto AS [Descripción Producto],
        p.precio_compra AS [Precio Compra],
        p.precio_venta AS [Precio Venta],
        d.cantidad AS [Cantidad Vendida],
        p.stock AS [Stock],
        v.fecha AS [Fecha Venta],
        v.total AS [Total Vendido],
        v.estado AS [Estado],
        COALESCE(de.TotalDescuento, 0) AS [Total Descuento],  -- Total descuento aplicado
        COALESCE(de.Usuario, 'N/A') AS [Usuario],              -- Usuario que aplicó el descuento
        COALESCE(de.FechaDescuento, v.fecha) AS [Fecha Descuento], -- Fecha del descuento
        COALESCE(de.Concepto, 'N/A') AS [Concepto],           -- Concepto del descuento
        'Activo' AS [Estado Venta]
    FROM venta v
    INNER JOIN detalleventa d ON v.num_documento = d.num_documento AND v.serie = d.serie
    INNER JOIN producto p ON d.cod_producto = p.cod_producto
    LEFT JOIN (
        -- Subconsulta para obtener el descuento más reciente por producto
        SELECT 
            CodProducto, 
            MAX(FechaDescuento) AS FechaDescuento,
            MAX(TotalDescuento) AS TotalDescuento,  -- Asume que quieres el descuento más alto asociado a la fecha
            MAX(Usuario) AS Usuario, 
            MAX(Concepto) AS Concepto
        FROM Descuentos
        GROUP BY CodProducto
    ) de ON d.cod_producto = de.CodProducto AND CONVERT(DATE, v.fecha) = CONVERT(DATE, de.FechaDescuento)
    WHERE v.num_documento = @NumeroDocumento;

    -- Contamos la cantidad de productos en la factura
    SELECT @CantidadProductos = COUNT(*)
    FROM detalleventa
    WHERE num_documento = @NumeroDocumento;
    
    -- Devolvemos la cantidad de productos
    SELECT @CantidadProductos AS [Cantidad de Productos];
END
GO
```

# [dbo].[sp_mostrar_ventas_con_detalles_por_fecha]

```sql
/****** Object:  StoredProcedure [dbo].[sp_mostrar_ventas_con_detalles_por_fecha]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_mostrar_ventas_con_detalles_por_fecha]
    @FechaInicio DATE,
    @FechaFin DATE
AS
BEGIN
    -- Mostrar ventas activas que no han sido eliminadas
    SELECT 
        v.num_documento AS [Número de Factura], 
        v.serie AS [Serie], 
        v.cod_cliente AS [Código Cliente], 
        v.dni AS [DNI], 
        d.cod_producto AS [Código Producto],
        p.nombre AS [Nombre Producto],
        p.descripcion_producto AS [Descripción Producto],
        p.precio_compra AS [Precio Compra],
        p.precio_venta AS [Precio Venta],
        d.cantidad AS [Cantidad Vendida],
        p.stock AS [Stock],
        v.fecha AS [Fecha Venta], 
        v.total AS [Total Vendido],
        v.estado AS [Estado],
        COALESCE(de.TotalDescuento, 0) AS [Total Descuento],  -- Total descuento aplicado
        COALESCE(de.Usuario, 'N/A') AS [Usuario],              -- Usuario que aplicó el descuento
        COALESCE(de.FechaDescuento, v.fecha) AS [Fecha Descuento], -- Fecha del descuento
        COALESCE(de.Concepto, 'N/A') AS [Concepto],           -- Concepto del descuento
        'Activo' AS [Estado Venta]
    FROM venta v
    INNER JOIN detalleventa d
        ON v.num_documento = d.num_documento 
        AND v.serie = d.serie
    INNER JOIN producto p
        ON d.cod_producto = p.cod_producto
    LEFT JOIN (
        -- Subconsulta para obtener el descuento más reciente por producto
        SELECT 
            CodProducto, 
            MAX(FechaDescuento) AS FechaDescuento, 
            MAX(TotalDescuento) AS TotalDescuento,  -- Asume que quieres el descuento más alto asociado a la fecha
            MAX(Usuario) AS Usuario, 
            MAX(Concepto) AS Concepto
        FROM Descuentos
        GROUP BY CodProducto
    ) de
        ON d.cod_producto = de.CodProducto 
        AND CONVERT(DATE, v.fecha) = CONVERT(DATE, de.FechaDescuento)
    -- Excluir las ventas que ya están en ventas_eliminadas
    WHERE v.fecha BETWEEN @FechaInicio AND @FechaFin
    AND NOT EXISTS (
        SELECT 1 
        FROM ventas_eliminadas ve
        WHERE ve.num_documento = v.num_documento
    )

    UNION ALL

    -- Mostrar ventas eliminadas con detalles de producto
    SELECT 
        ve.num_documento AS [Número de Factura], 
        ve.serie AS [Serie], 
        ve.cod_cliente AS [Código Cliente], 
        ve.dni AS [DNI], 
        ve.cod_producto AS [Código Producto],
        ve.nombre_producto AS [Nombre Producto],
        ve.descripcion_producto AS [Descripción Producto],
        ve.precio_compra AS [Precio Compra],
        ve.precio_venta AS [Precio Venta],
        ve.cantidad AS [Cantidad Vendida],
        NULL AS [Stock],  -- No es necesario mostrar stock ya que la venta ha sido eliminada
        ve.fecha AS [Fecha Venta], 
        ve.total AS [Total Vendido],
        ve.estado AS [Estado],
        NULL AS [Total Descuento], -- No aplica en ventas eliminadas
        NULL AS [Usuario],          -- No aplica en ventas eliminadas
        NULL AS [Fecha Descuento],  -- No aplica en ventas eliminadas
        NULL AS [Concepto],          -- No aplica en ventas eliminadas
        'Eliminado' AS [Estado Venta]
    FROM ventas_eliminadas ve
    WHERE ve.fecha_eliminacion BETWEEN @FechaInicio AND @FechaFin

    ORDER BY [Fecha Venta];
END
GO
```

# [dbo].[sp_obtener_descuento_por_num_documento]

```sql
/****** Object:  StoredProcedure [dbo].[sp_obtener_descuento_por_num_documento]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_obtener_descuento_por_num_documento]
    @num_documento VARCHAR(20)
AS
BEGIN
    SELECT totaldescuento, fecha, num_documento
    FROM TotalDescuento
    WHERE num_documento = @num_documento;
END;
GO
```

# [dbo].[sp_obtener_productos_por_proveedor]

```sql
/****** Object:  StoredProcedure [dbo].[sp_obtener_productos_por_proveedor]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_obtener_productos_por_proveedor]
    @id_proveedor INT
AS
BEGIN
    SELECT 
        p.nombre AS NombreProducto,
        p.Descripcion_producto AS DescripcionProducto,
        p.precio_compra AS PrecioCompra,
        prov.nombre AS NombreProveedor,
        prov.direccion AS DireccionProveedor,
        prov.telefono AS TelefonoProveedor,
        prov.email AS EmailProveedor,
        p.fecha_registro AS FechaRegistro
    FROM 
        Producto p
    INNER JOIN 
        Proveedores prov ON p.id_proveedor = prov.id_proveedor 
    WHERE 
        p.id_proveedor = @id_proveedor
        AND p.estado = '1' -- Solo selecciona productos activos
END
GO
```

# [dbo].[sp_obtener_total_descuento_por_fecha]

```sql
/****** Object:  StoredProcedure [dbo].[sp_obtener_total_descuento_por_fecha]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_obtener_total_descuento_por_fecha]
    @fechaInicio DATETIME,
    @fechaFin DATETIME
AS
BEGIN
    SELECT SUM(totaldescuento) AS totalDescuento
    FROM TotalDescuento
    WHERE fecha BETWEEN @fechaInicio AND @fechaFin;
END;
GO
```

# [dbo].[validar_usuario]

```sql
/****** Object:  StoredProcedure [dbo].[validar_usuario]    Script Date: 08/04/2026 22:24:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[validar_usuario]
    @usuario NVARCHAR(40),
    @contraseña NVARCHAR(40)
AS
BEGIN
    SET NOCOUNT ON;

    IF NOT EXISTS (SELECT 1 FROM usuario WHERE usuario = @usuario)
    BEGIN
        SELECT NULL AS cod_usuario, -1 AS ErrorCode; -- Usuario no existe
        RETURN;
    END

    IF NOT EXISTS (SELECT 1 FROM usuario WHERE usuario = @usuario AND contraseña = @contraseña)
    BEGIN
        SELECT NULL AS cod_usuario, -2 AS ErrorCode; -- Contraseña incorrecta
        RETURN;
    END

    SELECT cod_usuario, 0 AS ErrorCode, tipo_usuario, estado -- Usuario y contraseña válidos
    FROM usuario
    WHERE usuario = @usuario AND contraseña = @contraseña;
END;
GO
```
